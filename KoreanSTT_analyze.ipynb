```python
import pandas as pd
from glob import glob
```

```python
file_list=glob('cerwer/*')
```

```python
# 음원 파일의 길이가 담긴 데이터프레임 불러오기
duration=[]
for file in file_list:
    if "duration" in file:
        duration.append(file)

duration_df=pd.DataFrame()
for d in duration:
    tmp=pd.read_csv(d)
    duration_df=pd.concat([duration_df, tmp])
```

```python
def analyze(duration_df):
    api=input("returnzero, naver, azure, google, whisper 중 하나 입력: ")
    target=[]
    for file in file_list:
        if api in file:
            target.append(file)
            
    zhengli=[]
    for t in target:
        api=t.split('_')[-1].split('.csv')[0]
        dataset=t.split('cw_')[1].split('_'+api)[0]
        df=pd.read_csv(t)
        
        # 인식 안된 데이터 제외 및 카운트
        null_df=df[((df['recognized_text'].isnull())|(df['recognized_text']=='error'))]
        df=df[~((df['recognized_text'].isnull())|(df['recognized_text']=='error'))]
        
        # 문자열일경우
        df['execution_time']=df['execution_time'].astype('float64')
        
        null_cnt=len(null_df)
        df=pd.merge(df, duration_df, on='file_name', how='left')
        mt_average_length = df['modified_text'].str.len().mean()
        rt_average_length = df['recognized_text'].str.len().mean()
        time_average=df['execution_time'].mean()
        cer_average=df['cer'].mean()
        wer_average=df['wer'].mean()
        duration_average=df['duration'].mean()
        zhengli.append([api, dataset,null_cnt, mt_average_length, rt_average_length, time_average,cer_average,wer_average, duration_average])
    zhengli=pd.DataFrame(zhengli)
    zhengli.columns=['api','dataset','null_cnt', 'mt_average_length', 'rt_average_length', 'time_average','cer_average','wer_average', 'duration_average']
    return zhengli
```

```python
returnzero=analyze(duration_df)
```

```python
def total_analyze(duration_df):
    total=pd.DataFrame()
    api_list=['returnzero','naver','azure','google','whisper']
    for api in api_list:
        target=[]
        for file in file_list:
            if api in file:
                target.append(file)

        zhengli=[]
        for t in target:
            api=t.split('_')[-1].split('.csv')[0]
            dataset=t.split('cw_')[1].split('_'+api)[0]
            df=pd.read_csv(t)

            # 인식 안된 데이터 제외 및 카운트
            null_df=df[((df['recognized_text'].isnull())|(df['recognized_text']=='error'))]
            df=df[~((df['recognized_text'].isnull())|(df['recognized_text']=='error'))]

            # 문자열일경우
            df['execution_time']=df['execution_time'].astype('float64')

            null_cnt=len(null_df)
            df=pd.merge(df, duration_df, on='file_name', how='left')
            mt_average_length = df['modified_text'].str.len().mean()
            rt_average_length = df['recognized_text'].str.len().mean()
            time_average=df['execution_time'].mean()
            cer_average=df['cer'].mean()
            wer_average=df['wer'].mean()
            duration_average=df['duration'].mean()
            zhengli.append([api, dataset,null_cnt, mt_average_length, rt_average_length, time_average,cer_average,wer_average, duration_average])
        zhengli=pd.DataFrame(zhengli)
        zhengli.columns=['api','dataset','null_cnt', 'mt_average_length', 'rt_average_length', 'time_average','cer_average','wer_average', 'duration_average']
        zhengli=zhengli.groupby(['api'])[['null_cnt','time_average','cer_average','wer_average']].mean()
        zhengli.reset_index(inplace=True)
        total=pd.concat([total, zhengli])
    return total
```

```python
total=total_analyze(duration_df)
```
